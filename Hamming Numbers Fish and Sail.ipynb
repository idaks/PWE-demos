```python
%%clingo -l hamming_gv_rules --donot-display_input --save_meta_data_to md --saveto fish_out

% "One-loop variant"

% schema hamming(Output, Input, Factor)
% Generates Hamming numbers by simple multiplication:
%    Output := Input * Factor
% The third argument (Factor) also acts as a *provenance annotation*,
% which can be used to create labeled edges: Input --Factor--> Output

% Getting started - add Hamming number 1:
hamming(1, 1, 1). 

% If X is a Hamming number, so is Y = F*X (also record F in 3rd "annotation" column):
% limit set to Y < 1000
hamming(Y,X, 2) :- hamming(X,_, _), Y = 2*X, Y < 1000.
hamming(Y,X, 3) :- hamming(X,_, _), Y = 3*X, Y < 1000.
hamming(Y,X, 5) :- hamming(X,_, _), Y = 5*X, Y < 1000.
```

```python
%%clingo -l hamming_gv_rules --donot-display_input --save_meta_data_to md --saveto sail_out

%% Three Loop Variant

% Below, we need the sets {1,2}, {1,2,3}, {1,2,3,5}, so let's define them: 
in_12(1).
in_12(2).

in_123(X) :- in_12(X).
in_123(3).

in_1235(X) :- in_123(X).
in_1235(5).

% schema hamming(Output, Input, Factor)
% Generates Hamming numbers by simple multiplication:
%    Output := Input * Factor
% The third argument (Factor) also acts as a *provenance annotation*,
% which can be used to create labeled edges: Input --Factor--> Output

% The first rule (rather: fact) gets us started and adds "1" as a Hamming number.
% The second rule generates multiples of 2 from numbers that have 1 or 2 as a factor.
% The third rule generates multiples of 3 from numbers that have 1,2, or 3 as a factor.
% The fourth rule generates multiples of 5 from numbers that have 1,2, 3, or 5 as a factor.
% limit set to Y < 1000
hamming(1,1, 1).
hamming(Y,X, 2) :- hamming(X,_, F), in_12(F),   Y = 2*X , Y<1000.
hamming(Y,X, 3) :- hamming(X,_, F), in_123(F),  Y = 3*X , Y<1000.
hamming(Y,X, 5) :- hamming(X,_, F), in_1235(F), Y = 5*X , Y<1000.
```

Some Basic Regular Path Queries

```python
%%clingo --donot-display_input --donot-run -lci node_path_defs --donot-run

node(K) :- hamming(K,_,_).
node(K) :- hamming(_,K,_).

path(A,B) :- hamming(A,B,_).
path(A,B) :- hamming(A,C,_), path(C,B).
```

```python
%%clingo --donot-display_input --donot-run -lci star_def

star_k(2). star_k(3). star_k(5).

star(A,A,K) :- node(A), star_k(K).
star(A,B,K) :- hamming(A,B,K), star_k(K).
star(A,B,K) :- hamming(A,C,K), star(C,B,K), star_k(K).
```

```python
%%clingo --donot-display_input --donot-run -lci two_three_five_star_rule_check

% 2*.3*.5*
two_three_five_star_path(A,B) :- star(A,C,5), star(C,D,3), star(D,B,2).

%:- path(A,B), not two_three_five_star_path(A,B).

#show two_three_five_star_path/2.
```
